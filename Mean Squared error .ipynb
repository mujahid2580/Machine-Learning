# CALCULATING MSE FOR SAMPLE DATA

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.metrics import mean_squared_error
```

```python
# Generate n evenly spaced values from zero rasiand to 2 pi radians
n= 200
x= np.linspace(0,2*np.pi,n)
sine_values= np.sin(x)
```

```python
plt.plot(x,sine_values)
```

```python
# add some noise to the sin curve
noise= 0.5
noisy_sine_values= sine_values+np.random.uniform(-noise,noise,n)
# plotting noise data
plt.plot(x,noisy_sine_values,color='r')
plt.plot(x,sine_values, linewidth=3,color='Black')
```

```python
# calculate mean squared error usinf=g th equation
error_value= (1/n)*sum(np.power(sine_values - noisy_sine_values,2))
error_value
```

```python
mean_squared_error(sine_values,noisy_sine_values)
```
