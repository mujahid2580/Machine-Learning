# Gradient Descent
    Gradient is an optimization algorithm to minimize an function
    It is impossible to visualize gradient in a space consisting of more than 3 dimensions
    

    . Optimization Algorithm to minimize a function
    . Iteratively move in the direction of steepest descent
    . Steepest descent
        Largest Negative Gradient

### Gradient descent in Linear Regression
    .Model Hypothesis --> Linear Fit
                            y=m*X + c
    .Cost function --> Mean squared error
    
\begin{equation*}
MSE = \frac{1}{n}\sum_{i=1}^n|y-y_i|^2
\end{equation*}

### Gradient descent function is implemented with three functions they are
    .Update function
    .Error function
    .Gradient descent function
    
#### Update function:
    .Purpose:
        1.Calculate Gradient at current position
        2.update values of weights such that gets closer to their final values
     
     . Parameters
         1.Current values of weights
         2.Attribute values(x)
         3.Target values(t)
         4.Learning Rate
     
     .Return:
     .Updated values of weights
