```python
import requests
```

```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.goibibo.com/hotels/intl-hotels-in-bali-ct/"
```

```python
headers = {
    'User-Agent':"Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36(KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36"
}
```

```python
response = requests.request("GET", url,headers=headers)
```

```python
data = BeautifulSoup(response.text, 'html.parser')
print(data)
```

```python
cards_data = data.find_all('div',attrs={'class',"HotelCardstyles__WrapperSectionMetaDiv-sc-1s80tyk-2 iytEUf"})
```

```python
print('Total number of cards found:', len(cards_data))
```

```python
for card in cards_data:
    print(card)
```

```python
for card in cards_data:
    hotel_name = card.find('a')
    #to get the price of room
    room_price = card.find('p', attrs={'class': "HotelCardstyles__CurrentPrice-sc-1s80tyk-27 czKsrL"})
    print(hotel_name.text, room_price.text)
```

```python
for card in cards_data:
    ratings= card.find('span', attrs={'class':"ReviewAndRatingsstyles__AverageReviewText-sc-1nxmeoo-8 MUGDq"})
    address = card.find('span',attrs = {'class':"PersuasionHoverTextstyles__HoverTargetWrapperDiv-sc-1c06rw1-2 eGhSZO"})
    print(ratings.text,address.text)
```

```python
scraped_data = []
for card in cards_data:
    #initialize the dictionary
    card_details = {}
    
    #get the hotel name
    hotel_name= card.find('a')
    #get the room price
    room_price = card.find('p',attrs={'class':"HotelCardstyles__CurrentPrice-sc-1s80tyk-27 czKsrL"})
    #Hotel ratings
    ratings = card.find('span',attrs={'class':"ReviewAndRatingsstyles__AverageReviewText-sc-1nxmeoo-8 MUGDq"})
    #address of the hotel
    address = card.find('span',attrs={'class':"PersuasionHoverTextstyles__HoverTargetWrapperDiv-sc-1c06rw1-2 eGhSZO"})
    #add data to the dictionary
    card_details['Hotel_name'] = hotel_name.text
    card_details['Room_price'] = room_price.text
    card_details['Ratings']= ratings.text
    card_details['Address']=address.text
    
    #append the scraped data to the list
    scraped_data.append(card_details)
#create a data frmae from the list of dictionaries
df = pd.DataFrame.from_dict(scraped_data)
#save the scraped data in csv file
#df.to_csv('hotels data.csv',index=False)
```
